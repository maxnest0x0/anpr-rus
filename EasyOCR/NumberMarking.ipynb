```python
import cv2
import os

from google.colab.patches import cv2_imshow
```

```python
# Функция для маркировки номеров
def number_marking(folder_path: str, output_folder_path: str) -> None:
    images_names = os.listdir(folder_path)
    count = 0

    for image_name in images_names:
        count += 1

        image_path = os.path.join(folder_path, image_name)
        image = cv2.imread(image_path)
        cv2_imshow(image)

        new_image_name = input() + '.bmp'
        output_path = os.path.join(output_folder_path, new_image_name)

        cv2.imwrite(output_path, image)
        print(f'Обработано изображение №{count}')

    print('Конец!!!')
```

```python
# Путь к папке с изображениями номеров
folder_path = '/content/CropCarNumbers/NumBase'
# Путь к папке, куда будут загружены переименованные изображения
output_folder_path = '/content/CropCarNumbersWithMarking'

# Начало маркировки номеров
number_marking(folder_path, output_folder_path)
```

```python
# Количество переименнованых изображенией
print(len(os.listdir(output_folder_path)))
```

```python
def save_on_drive(folder_path: str, save_path: str):
    images_names = os.listdir(folder_path)

    for image_name in images_names:
        image_path = os.path.join(folder_path, image_name)
        image = cv2.imread(image_path)

        output_path = os.path.join(save_path, image_name)
        cv2.imwrite(output_path, image)

    print('Конец!!!')
```

```python
folder_path = '/content/CropCarNumbersWithMarking'
save_path = '/content/drive/MyDrive/Распознавание номеров/NumberWithMarking'

save_on_drive(folder_path, save_path)
```
